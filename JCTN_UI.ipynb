```python
#@title ## 2.1ã€å¢åˆ æ”¹æŸ¥mod/é€‰æ‹©é¢„ä¸‹è½½çš„modé›†åˆ
#@markdown ####é€‰æ‹©modé›†åˆ(å¦‚æœäº‘ç›˜æ²¡æ‰¾åˆ°é›†åˆåˆ™åœ¨æ ¹ç›®å½•åˆ›å»º)ï¼š##
mod_json_name = "model.json" #@param {type: "string"}
# åˆ›å»ºå¹¶å†™å…¥æ–‡æœ¬åˆ°æ–‡ä»¶
mod_json_dir=''

#é€’å½’ç›®å½•æŸ¥æ‰¾modé›†åˆ
def find_model_json(root_path,mod_json_name):
    for file_name in os.listdir(root_path):
        file_path = os.path.join(root_path, file_name)
        if os.path.isdir(file_path):
            res=find_model_json(file_path,mod_json_name)
            if res is not None:
              return res
        elif file_name == mod_json_name:
            print("æ‰¾åˆ°model.jsonæ–‡ä»¶ï¼š", file_path)
            return file_path

if 'is_load_pan' in globals() and is_load_pan=="å…è®¸":
  mod_json_dir=find_model_json('/content/drive/',mod_json_name)
  if mod_json_dir==None:
    with open(f'/content/drive/MyDrive/{mod_json_name}', 'w') as f:
      f.write('[]')
      mod_json_dir=f'/content/drive/MyDrive/{mod_json_name}'
else:
  mod_json_dir=f'/content/{mod_json_name}'

import json
import pandas as pd
# drive.mount('/content/drive')
from IPython.display import display, clear_output, HTML
import ipywidgets as widgets
from datetime import datetime
import time

custom_css = """
<style>
    .custom-button {
        padding: 10px 25px;
    }
    .custom-button.save {
        background-color: #67C23A;
        float:right
    }
    .custom-button.add {
        background-color: #409EFF;
    }
    .custom-button.delete {
        background-color: #F56C6C;
    }
    .custom-button.update {
        background-color: #E6A23C;
    }
</style>
"""
display(HTML(custom_css))

# åŠ è½½JSONæ•°æ®ä¸ºPythonå¯¹è±¡
with open(mod_json_dir, 'r') as f:
    data = json.load(f)

# å°†JSONæ•°æ®è½¬æ¢ä¸ºPandas DataFrame
df = pd.DataFrame(data)

# æ˜¾ç¤ºè¡¨æ ¼
def show_table():
    clear_output()
    styled_df = df.style.set_properties(**{'text-align': 'center'})

    # è®¾ç½®åˆ—å®½åº¦å’Œå±…ä¸­å¯¹é½
    styled_df.set_table_styles([{'selector': 'th', 'props': [('max-width', '150px'), ('text-align', 'center')]}])

    # è¾“å‡ºDataFrame
    display(styled_df)
    display(widgets.HBox([input_name, input_type, input_download_link, input_size]))
    display(widgets.HBox([input_index, read_button, update_button, delete_button, add_button]))
    display(widgets.HBox([save_button]))

# æ·»åŠ ä¸€è¡Œæ•°æ®
def add_row(button):
  try:
    if not input_download_link.value:
      display(HTML(f"<p style='color: red;'>ä¸‹è½½é“¾æ¥ä¸ºå¿…å¡«é¡¹</p>"))
      return
    new_row = {
        "name": input_name.value.strip() or 'æœªå‘½å',
        "type": 'æœªå®šä¹‰' if input_type.value == 'ä¸é€‰åˆ™æ ¹æ®ä¸‹è½½å†…å®¹å†³å®š' else input_type.value,
        "downloadLink": input_download_link.value.strip(), # ä¿®æ”¹è¿™é‡Œçš„é”®å
        "size": input_size.value,
        "lastChangeDate": datetime.now().strftime('%Y-%m-%d')
    }
    global df
    df = df.append(new_row, ignore_index=True)
    lock_buttons()
    show_table()
    clear_input()
    display(HTML("<p id='add-msg' style='color: green;'>æ–°å¢{}æˆåŠŸ</p>".format(new_row)))
    time.sleep(1)
    display(HTML("<script>document.getElementById('add-msg').remove()</script>"))
  except Exception as e:
    display(HTML(f"<p style='color: red;'>æ–°å¢å‡ºé”™ï¼š{e}</p>"))

# åˆ é™¤ä¸€è¡Œæ•°æ®
def delete_row(button):
  try:
    index = int(input_index.value)
    global df
    df = df.drop(index)
    lock_buttons()
    show_table()
    display(HTML("<p id='delete-msg' style='color: green;'>åˆ é™¤{}æˆåŠŸ</p>".format(index)))
    time.sleep(1)
    display(HTML("<script>document.getElementById('delete-msg').remove()</script>"))
  except Exception as e:
    display(HTML(f"<p style='color: red;'>åˆ é™¤å‡ºé”™ï¼š{e}</p>"))

#è¯»å–ä¸€è¡Œæ•°æ®
def read_row(button):
    try:
        index = int(input_index.value)
        input_name.value = df.at[index, 'name']
        input_type.value ='ä¸é€‰åˆ™æ ¹æ®ä¸‹è½½å†…å®¹å†³å®š' if df.at[index, 'type']=='æœªå®šä¹‰' else df.at[index, 'type']
        input_download_link.value = df.at[index, 'downloadLink']
        input_size.value = df.at[index, 'size']
        display(HTML(f"<p id='read-msg' style='color: green;'>è¯»å–{index}æˆåŠŸ</p>"))
        time.sleep(1)
        display(HTML("<script>document.getElementById('read-msg').remove()</script>"))
    except Exception as e:
        display(HTML(f"<p style='color: red;'>è¯»å–å‡ºé”™ï¼š{e}</p>"))
# ä¿®æ”¹ä¸€è¡Œæ•°æ®
def update_row(button):
    try:
      index = int(input_index.value)
      if input_download_link.value:
        df.at[index, 'downloadLink'] = input_download_link.value.strip()
      else:
        display(HTML(f"<p style='color: red;'>ä¸‹è½½é“¾æ¥ä¸ºå¿…å¡«é¡¹{e}</p>"))
        return
      df.at[index, 'name'] = input_name.value.strip() or 'æœªå‘½å'
      df.at[index, 'type'] = 'æœªå®šä¹‰' if input_type.value == 'ä¸é€‰åˆ™æ ¹æ®ä¸‹è½½å†…å®¹å†³å®š' else input_type.value
      df.at[index, 'size'] = input_size.value
      df.at[index, 'lastChangeDate'] = datetime.now().strftime('%Y-%m-%d')
      lock_buttons()
      clear_input()
      show_table()
      display(HTML("<p id='update-msg' style='color: green;'>ä¿®æ”¹{}æˆåŠŸ</p>".format(index)))
      time.sleep(1)
      display(HTML("<script>document.getElementById('update-msg').remove()</script>"))
    except Exception as e:
      display(HTML(f"<p style='color: red;'>ä¿®æ”¹å‡ºé”™ï¼š{e}</p>"))

#æ¸…ç©ºè¾“å…¥æ¡†
def clear_input():
    try:
        input_name.value = ''
        input_type.value = 'lora'
        input_download_link.value = ''
        input_size.value = '144MB'
    except Exception as e:
        display(HTML(f"<p style='color: red;'>é‡ç½®è¾“å…¥æ¡†å‡ºé”™ï¼š{e}</p>"))

#ä¿å­˜
def save_row(button):
    global df
    try:
        with open(mod_json_dir, 'w') as Wfile:
            json=df.to_json(orient='records')
            Wfile.write(json)
            display(HTML("<p id='save-msg' style='color: green;'>ä¿å­˜æˆåŠŸ</p>"))
            time.sleep(1)
            display(HTML("<script>document.getElementById('save-msg').remove()</script>"))
    except Exception as e:
        display(HTML(f"<p style='color: red;'>ä¿å­˜å‡ºé”™ï¼š{e}</p>"))



# å®šä¹‰è¾“å…¥æ¡†å’ŒæŒ‰é’®
input_name = widgets.Text(description='åå­—',placeholder='ä¸å¡«åˆ™æ ¹æ®ä¸‹è½½å†…å®¹å†³å®š')
input_type = widgets.Dropdown(
    options=['ä¸é€‰åˆ™æ ¹æ®ä¸‹è½½å†…å®¹å†³å®š','lora', 'checkpoint'],
    value='ä¸é€‰åˆ™æ ¹æ®ä¸‹è½½å†…å®¹å†³å®š',
    description='ç±»å‹:',
    disabled=False,
)
input_download_link = widgets.Text(description='ä¸‹è½½é“¾æ¥',placeholder='å¿…å¡«ï¼šCç«™æˆ–è€…huggingface') # ä¿®æ”¹è¿™é‡Œçš„æè¿°
input_size = widgets.Text(description='å¤§å°',placeholder='é€‰å¡«')
input_index = widgets.Text(description='modç¼–å·',placeholder='ä¸æ–°å¢æ— å…³ï¼Œç”¨äºè¯»æ”¹åˆ mod',layout=widgets.Layout(margin='10px'))
add_button = widgets.Button(description='æ–°å¢mod', layout=widgets.Layout(margin='10px 180px'), button_style='info')
delete_button = widgets.Button(description='åˆ é™¤mod', layout=widgets.Layout(margin='10px 10px'), button_style='danger')
update_button = widgets.Button(description='æ›´æ–°mod', layout=widgets.Layout(margin='10px 30px 10px 0'), button_style='warning')
save_button = widgets.Button(description='ä¿å­˜', layout=widgets.Layout(margin='20px auto'), button_style='success')
read_button = widgets.Button(description='è¯»å–mod', layout=widgets.Layout(margin='10px 0px 10px 30px'), button_style='primary')


#èŠ‚æµ
def lock_buttons():
    # add_button.disabled = True
    delete_button.disabled = True
    update_button.disabled = True

def unlock_buttons():
    # add_button.disabled = False
    delete_button.disabled = False
    update_button.disabled = False

def on_input_change(change):
  if change.new:
    # len(change.new)
    if change.new==change.owner.value:
      if len(change.new)==1:
        lock_buttons()
        time.sleep(0.5)
        unlock_buttons()
  else:
    lock_buttons()


# ä¸ºè¾“å…¥æ¡†æ·»åŠ è¾“å…¥äº‹ä»¶
# input_name.observe(on_input_change, names='value')
# input_type.observe(on_input_change, names='value')
# input_download_link.observe(on_input_change, names='value')
# input_size.observe(on_input_change, names='value')
input_index.observe(on_input_change, names='value')

add_button.on_click(add_row)
delete_button.on_click(delete_row)
update_button.on_click(update_row)
save_button.on_click(save_row)
read_button.on_click(read_row)

# æ˜¾ç¤ºè¡¨æ ¼å’Œäº¤äº’å¼æŒ‰é’®
lock_buttons()
show_table()
```

```python
#@title ## 2.2ã€ä¸‹è½½æ‰€æœ‰mod
#@markdown ####ï¼ˆå¯é€‰ï¼‰åªä¸‹ä¸€éƒ¨åˆ†modå°±å¡«ï¼ˆå¤åˆ¶è¡¨æ ¼modåæŒ‰â€œä¸â€œå­—åˆ†å‰²ï¼‰ï¼š##
mod_name = "" #@param {type: "string"}
sd="s"+"t"+"a"+"b"+"l"+"e"+"-"+"d"+"i"+"f"+"f"+"u"+"s"+"i"+"o"+"n"+"-"+"w"+"e"+"b"+"u"+"i"

import json
import re
import subprocess
import shutil
import os
from concurrent.futures import ThreadPoolExecutor

def swap(arr,l,r):
  temp=arr[l]
  arr[l]=arr[r]
  arr[r]=temp

def checkRelateMod(oldMod, newMod):
  try:
    return oldMod and newMod and oldMod['downloadLink'] == newMod['downloadLink']
  except Exception as e:
    print(oldMod)
    print(e)

def dealRelateMod(oldStartMod, newStartMod):
  if oldStartMod['name'] != newStartMod['name']:
    for file in os.listdir(f'/content/{sd}/models/Lora/'):
      if os.path.splitext(file)[0] == oldStartMod['name']:  # å¿½ç•¥æ‰©å±•åè¿›è¡Œåˆ¤æ–­
          os.rename(f"/content/{sd}/models/Lora/{file}",f"/content/{sd}/models/Lora/{newStartMod['name']}{os.path.splitext(file)[1]}")
          print(f"{oldStartMod['name']}é‡å‘½åæˆåŠŸ")
          return
    for file in os.listdir(f"/content/{sd}/models/Stable-diffusion/"):
      if os.path.splitext(file)[0] == oldStartMod['name']:  # å¿½ç•¥æ‰©å±•åè¿›è¡Œåˆ¤æ–­
          os.rename(f"/content/{sd}/models/Stable-diffusion/{file}",f"/content/{sd}/models/Stable-diffusion/{newStartMod['name']}{os.path.splitext(file)[1]}")
          print(f"{oldStartMod['name']}é‡å‘½åæˆåŠŸ")
          return

  else:
    print(f"{newStartMod['name']}å·²å­˜åœ¨ï¼Œå°†å¿½è§†")

def deleteMod(oldDeMod):
  for file in os.listdir(f'/content/{sd}/models/Lora/'):
          if os.path.splitext(file)[0] == oldDeMod:  # å¿½ç•¥æ‰©å±•åè¿›è¡Œåˆ¤æ–­
              os.remove(f"/content/{sd}/models/Lora/{file}")
              print(f"{oldDeMod}åˆ é™¤æˆåŠŸ")
              return
  for file in os.listdir(f"/content/{sd}/models/Stable-diffusion/"):
    if os.path.splitext(file)[0] == oldDeMod:  # å¿½ç•¥æ‰©å±•åè¿›è¡Œåˆ¤æ–­
        os.remove(f"/content/{sd}/models/Stable-diffusion/{file}")
        print(f"{oldDeMod}åˆ é™¤æˆåŠŸ")
        return

def LCS(oldCo, newCo):
  oldStartIdx = 0   # æ—§å‰æŒ‡é’ˆ
  newStartIdx = 0   # æ–°å‰æŒ‡é’ˆ
  oldEndIdx = len(oldCo) - 1  # æ—§åæŒ‡é’ˆ
  newEndIdx = len(newCo) - 1  # æ–°åæŒ‡é’ˆ
  oldStartMod = oldCo[oldStartIdx]  # æ—§å‰Mod
  oldEndMod = oldCo[oldEndIdx]  # æ—§åMod
  newStartMod = newCo[newStartIdx]  # æ–°å‰Mod
  newEndMod = newCo[newEndIdx]  # æ–°åMod
  linkMap = {}
  while oldStartIdx <= oldEndIdx and newStartIdx <= newEndIdx:
    # æ–°å‰æ—§å‰
    if checkRelateMod(oldStartMod, newStartMod):
      dealRelateMod(oldStartMod, newStartMod)
      oldStartIdx += 1
      newStartIdx += 1
      if newStartIdx > newEndIdx or oldStartIdx > oldEndIdx:
        break
      oldStartMod = oldCo[oldStartIdx]
      newStartMod = newCo[newStartIdx]
    # æ–°åæ—§å
    elif checkRelateMod(oldEndMod, newEndMod):
      dealRelateMod(oldEndMod, newEndMod)
      oldEndIdx -= 1
      newEndIdx -= 1
      if newStartIdx > newEndIdx or oldStartIdx > oldEndIdx:
        break
      oldEndMod = oldCo[oldEndIdx]
      newEndMod = newCo[newEndIdx]
    # æ–°åæ—§å‰
    elif checkRelateMod(oldStartMod, newEndMod):
      dealRelateMod(oldStartMod, newEndMod)
      oldStartIdx += 1
      newEndIdx -= 1
      if newStartIdx > newEndIdx or oldStartIdx > oldEndIdx:
        break
      oldStartMod = oldCo[oldStartIdx]
      newEndMod = newCo[newEndIdx]
    # æ–°å‰æ—§å
    elif checkRelateMod(oldEndMod, newStartMod):
      dealRelateMod(oldEndMod, newStartMod)
      oldEndIdx -= 1
      newStartIdx += 1
      if newStartIdx > newEndIdx or oldStartIdx > oldEndIdx:
        break
      oldEndMod = oldCo[oldEndIdx]
      newStartMod = newCo[newStartIdx]
    # å››ç§å‡æœªæ‰¾åˆ°
    else:
      # if not linkMap:
      linkMap = {}
    # ä» oldStartIdx å¼€å§‹ï¼Œåˆ°oldEndIdxç»“æŸï¼Œåˆ›å»ºlinkMapæ˜ å°„å¯¹è±¡
      for i in range(oldStartIdx, oldEndIdx+1):
        downloadLink = oldCo[i]['downloadLink']
        if downloadLink is not None:
          linkMap[downloadLink] = i
      idxInOld = linkMap.get(newStartMod['downloadLink'], None)
      if idxInOld is None:
        print(f"{newStartMod['name']}ä¸å­˜åœ¨äºæ—§mod,{newStartMod['downloadLink']}å°†æ·»åŠ åˆ°ä¸‹è½½ä»»åŠ¡é˜Ÿåˆ—")
        content.append(newStartMod)

      else:
        dealRelateMod(oldCo[idxInOld], newStartMod)
        swap(oldCo,idxInOld,oldEndIdx)
        oldEndIdx-=1
      # æŒ‡é’ˆä¸‹ç§»ï¼Œç§»åŠ¨æ–°çš„å¤´
      newStartIdx += 1
      if newStartIdx > newEndIdx:
          break
      newStartMod = newCo[newStartIdx]
      oldEndMod=oldCo[oldEndIdx]
      # oldStartIdx += 1
  print('new',newStartIdx,newEndIdx)
  print('old',oldStartIdx,oldEndIdx)
  if newStartIdx <= newEndIdx:
  # newè¿™é‡Œè¿˜æœ‰å‰©ä½™Modæ²¡æœ‰å¤„ç†
    for i in range(newStartIdx, newEndIdx + 1):
      print(f"{newCo[i]['name']}ä¸ºæ–°ä¸”ä¸é‡å¤modï¼Œå°†æ·»åŠ åˆ°ä¸‹è½½ä»»åŠ¡é˜Ÿåˆ—")
      content.append(newCo[i])

  i = oldStartIdx
  while i <= oldEndIdx:
      print(f"{oldCo[i]['name']}ä¸ºæ—§ä¸é‡å¤modï¼Œå°†åˆ é™¤")
      deleteMod(oldCo[i]['name'])
      del oldCo[i]
      oldEndIdx -= 1 # åˆ é™¤å…ƒç´ åï¼Œä¿®æ”¹ oldEndIdx çš„å€¼

# LCS(oldArr,oldArr)
with open(mod_json_dir, 'r') as modelFile:
  preContent=json.loads(modelFile.read())

#æ›´æ–°é˜¶æ®µ
if 'oldCo' in globals() and len(oldCo)>0:
  content=[]
  print(preContent)
  LCS(oldCo, preContent)
  print(content)
#åˆå§‹åŒ–é˜¶æ®µ
else:
  oldCo=[]
  content=preContent


#éƒ¨åˆ†ä¸‹è½½çš„æƒ…å†µ
selected_mods = [x.strip() for x in mod_name.split('ä¸') if x.strip()]


def get_file_size(file_path):
    try:
        file_size_bytes = os.path.getsize(file_path)
        for unit in ['B', 'KB', 'MB', 'GB', 'TB']:
            if file_size_bytes < 1024.0:
                break
            file_size_bytes /= 1024.0
        return f"{file_size_bytes:.2f} {unit}"
    except OSError as e:
        print(f"Error: {e}")
        return None

def get_civitai_file(str):
    match = re.search(r'[^/]+$', str.decode())
    if match:
      return match.group(0).split('\n')[0]
    else:
      print("æ— æ³•ä»è¾“å‡ºä¸­æå–æ–‡ä»¶å")

def move_model(source_path, target_path):
    try:
        shutil.move(source_path, target_path)
    except OSError as e:
        print(f"Error: {e}")

def download_file(item):
  # for item in content:
    if selected_mods and item['name'] not in selected_mods:
        return
    download_url=item['downloadLink']
    #huggingface
    match = re.search(r'/([^/]*)$', download_url)
    file_name = match.group(1)
    #å¼€å§‹ä¸‹è½½
    cmd = f"aria2c --console-log-level=error -c -x 16 -s 16 -k 1M {download_url} -d /content/{sd}/models/Lora" if '.' not in match.group(1) else f"aria2c --console-log-level=error -c -x 16 -s 16 -k 1M {download_url} -d /content/{sd}/models/Lora -o {file_name}"
    result = subprocess.run(cmd, shell=True, stderr=subprocess.PIPE, stdout=subprocess.PIPE)
    if result.returncode == 0:
      # åˆ¤æ–­æ˜¯civitaiè¿˜æ˜¯hugging faceè¿˜æ˜¯è‡ªå®šä¹‰
      #civitai
      if '.' not in file_name:
        file_name =get_civitai_file(result.stdout)
      #å¦‚æœå®šä¹‰äº†åå­—åˆ™é‡å‘½å
      if item['name']!='æœªå‘½å':
        match = re.search(r'\.([^\.]+)$', file_name)
        suffix = match.group(1)
        temp_name=item['name'].replace("/","")+'.'+suffix
        os.rename(f"/content/{sd}/models/Lora/{file_name}",f"/content/{sd}/models/Lora/{temp_name}")
        final_name =temp_name
      file_path = f"/content/{sd}/models/Lora/{final_name}"
      file_size = get_file_size(file_path)
      if file_size:
          print(f"{file_name}å·²ä¸‹è½½: {file_size}")
          #å¦‚æœä¸ºcheckpointåˆ™è¿›è¡Œç§»åŠ¨
          if 'GB'in file_size:
            source_path = f"/content/{sd}/models/Lora/{final_name}"
            target_path = f"/content/{sd}/models/Stable-diffusion/{final_name}"
            move_model(source_path, target_path)
            print('ç§»åŠ¨checkpoint--',final_name,f'åˆ°/content/{sd}/Stable-diffusion/æ–‡ä»¶å¤¹')
          print(file_name+'å·²ä¸‹è½½ï¼Œé‡å‘½åä¸ºï¼š'+final_name)
          oldCo.append(item)
      else:
          print(f"æ— æ³•è·å–{item['name']}å¤§å°")
    else:
        print(f"{item['name']}ä¸‹è½½å¤±è´¥,è¯·æ£€æŸ¥{item['downloadLink']}")

# download_file(item)
# ä½¿ç”¨çº¿ç¨‹æ± è¿›è¡Œä¸‹è½½
with ThreadPoolExecutor(max_workers=5) as executor:
    for item in content:
        executor.submit(download_file, item)
```
